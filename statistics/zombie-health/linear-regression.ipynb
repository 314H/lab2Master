```python
import pandas as pd

%matplotlib inline

df = pd.read_csv('zombie-health-age-days-recovery.csv')
```

```python
df.head(10)
```

```python
df.plot.scatter(x='age', y='days_recovery')
```

```python
from statsmodels.formula.api import ols

model = ols("age ~ days_recovery", data=df)
response = model.fit()
response.summary()
```

```python
y_pred = model.predict(df['age'])
print(y_pred)
```
